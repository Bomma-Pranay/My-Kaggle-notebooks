#### There is a problem here. Some values repeating in multiple rows. Finding all unique values by splitting with ',' 

```python
val = list(df['cuisine'])
val
```

```python
unique_cuisines = dict()
for i in val:
    tmp = i.split(',')
#     print(tmp)
    for j in tmp:
        if j not in unique_cuisines:
            unique_cuisines[j] = 1
        else:
            unique_cuisines[j] += 1
# unique_cuisines
# Sort descendingly
unique_cuisines = (sorted(unique_cuisines.items(), key=lambda item: item[1], reverse=True))
unique_cuisines

            
```

```python
# In unique_cuisines we see 166 'Desserts', that's the reason
# we see top-rated restaurants as Sweet shops.
```

```python
unique_cuisines_series = pd.Series(unique_cuisines)
unique_cuisines_series
```

# Let's checkout top-rated Biryani restaurants!!
